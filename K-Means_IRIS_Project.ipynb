```python
from sklearn.cluster import KMeans
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.preprocessing import scale
import sklearn.metrics as sm
from sklearn import datasets
```

```python
# Load the Iris dataset from sklearn.datasets
iris = datasets.load_iris()
print(iris.DESCR)
```

```python
# Create a DataFrame with features and labels
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df 
```

```python
# use only petal width and petal length as features
X = df[['petal length (cm)', 'petal width (cm)']]
X
```

```python
# Perform k-means clustering with 3 clusters (as there are 3 species of iris in the dataset)

inertia = []
for n_clusters in range(1, 11):
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    kmeans.fit_predict(X)
    inertia.append(kmeans.inertia_)

```

```python
# Plotting elbow method graph 
plt.plot(range(1, 11), inertia)
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal Number of Clusters')
plt.show()
```

### The resulting scatter plot will show how our data is clustered based on petal length and width. The elbow method graph will help us determine an optimal number of clusters for our data.
